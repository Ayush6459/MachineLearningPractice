```python
from bs4 import BeautifulSoup
import requests as req
```

```python
page=req.get('http://m.chia-anime.me/').text
```

```python
soup = BeautifulSoup(page,'lxml')
```

```python
#print(soup.prettify())
```

```python
for link in soup.find_all('tr',class_='episode'):
    print(link.a['href'])
    for x in link.find_all('span',class_=['series','name']):
        print(x.string)
    print('-------------------------------------------------')
    
    
    
    
```
